**Geoparsing**

Geoparsing merupakan salah satu metode ekstraksi data dari teks. Jika analisis Geocoding dapat melakukan seperti ini:

```
>> Geocode('Yogyakarta') = [-7.56218; 110.67921]
```

Maka Geoparsing dapat digunakan untuk melakukan ini:

```
>> Geoparse("Di sebelah timur Kota Magelang terdapat Kota Yogyakarta")

{
  Magelang : {
              lat:  -7.08712, 
              long: 110.472912
             };
  Yogyakarta: {
              lat:  -7.56218, 
              long: 110.67921
             };
}

```

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

from urllib import request
from geotext import GeoText

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
````

````python
# deteksi kota dengan Geotext

def detect_kota(text):
    try: 
        kota = GeoText(text)
        return list(kota.cities)
    except Exception:
        return 'exp'
````

````python
%%time

tweets['kota'] = tweets['tweet'].map(lambda t: detect_kota(t))
````
